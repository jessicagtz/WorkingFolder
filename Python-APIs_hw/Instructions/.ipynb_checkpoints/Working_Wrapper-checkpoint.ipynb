```python
cities = []
def newpoint():
    return uniform(-180, 180), uniform(-90, 90)

r_coordinates = (newpoint() for x in range(10))
for coordinate_pair in r_coordinates:
    #print (city)
    cities.append(coordinate_pair)
```
